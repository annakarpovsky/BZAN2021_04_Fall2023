    * Python keywords are off-limits.  See https://realpython.com/python-keywords/ for more info.
* Make names meaningful.

## Variable types

Each variable has a *type*.  You can use the `type()` function to figure out which type it is.

Common types
* Integers
* Floating point numbers (decimals)
* Strings
* Boolean:  True/False

```python
age = 83             #integer, used for integers
type(age)            
```

```python
temperature = 98.7   #float, used for real numbers (decimals)
type(temperature)
```

```python
type(color)         #we can use previously predefined variables
```

```python
name = "Luella"      #string, used for a collection of characters
type(name)
```

### Python uses "Duck typing:

You don't have to tell Python the data type of a variable. Python figures it out – this is commonly called *duck typing.*

## Changing the variable type

As we'll see in a minute, there are situations when Python does not give a variable the correct type.  In those cases you can use type conversion functions to manually change the type of a variable.
* `int()` changes a variable to an integer.
* `float()` changes a variable to a float.
* `str()` changes a variable to a string.

```python
num_string = "98.7"
num_float = float(num_string)
num_float
type(num_float)
```

Sometimes the results are not what you expect!  If you convert a float to an integer, Python just chops the decimal places off.

```python
int(98.7)
```

And if you try to convert a string to a float or an integer, you get an error.

```python
int("Go Eagles!")
```

```python
float("Go Eagles!")
```
